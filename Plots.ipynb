```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Superstore.xlsx')
df.head()
```

```python
sns.countplot(data=df, x='Segment')
```

```python
top_states = df['State'].value_counts().head(5)
```

```python
top_states
```

```python
sns.barplot(x=top_states.index, y=top_states.values)
plt.ylabel('Count')
plt.title('Top 5 States')
plt.show()
```

```python
top_cat = df['Sub-Category'].value_counts().head(5)
top_cat
```

```python
sns.barplot(x=top_cat.index, y=top_cat.values)
plt.ylabel('Count')
plt.title('Top 5 Sub Categories')
plt.show()
```

```python
sns.scatterplot(data=df, x='Profit', y="Sales")
```

```python
sns.barplot(data=df, x='Quantity', y="Discount")
```

```python
top_prod = df['Product Name'].value_counts().head(5)
top_prod
```

```python
sns.barplot(x=top_prod.index, y=top_prod.values)
plt.ylabel('Count')
plt.title('Top 5 Products')
plt.xticks(rotation=45)
plt.show()
```
